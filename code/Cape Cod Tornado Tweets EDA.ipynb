### Running EDA on Cape Cod Tornado Tweets

```python
import pandas as pd
```

```python
# Reading in .csv file into a dataframe

cape_cod_tornadoes = pd.read_csv('./datasets/cape_cod_tornado_12.csv')
```

```python
cape_cod_tornadoes_2 = pd.read_csv('./datasets/cape_cod_tornado_13.csv')
```

```python
# Combining the two .csv files into one dataframe

cape_cod_tornado_tweets = pd.concat([cape_cod_tornadoes, cape_cod_tornadoes_2])
```

```python
# Checking for duplicates

cape_cod_tornado_tweets.duplicated().sum()
```

```python
cape_cod_tornado_tweets.shape
```

```python
# Using this data as test. Opted not to remove duplicates, as this is technically unseen data.

cape_cod_tornado_tweets[cape_cod_tornado_tweets.duplicated()].sort_values(by = 'text')
```
